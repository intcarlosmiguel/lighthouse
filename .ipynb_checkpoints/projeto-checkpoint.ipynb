Agora há o principal problema do banco de dados, as classes desblanceadas. Esse é o maior problema desse banco de dados que é o desbalanceamento das classes. Temos várias formas de lidar com isso, a primeira que eu tentei foi a escolha aleatória dos dados, contudo não apresentou resultados satisfatórios, afinal eu poderia estar deletando dados importantes. Para evitar isso usaremos um algoritmo para fazermos a criação sintética de dados o **SMOTE**.

SMOTE é um algoritmo que criará dados cintéticos com o intuito de melhorar o balanceamento de classes. A ideia por trás dela consiste em criar observações intermediárias entre dados parecidos, ou seja, se no dataset existem 2 pessoas, uma com altura 1,80 m e pesando 78 kg, a outra com 1,82 m e pesando 79 kg; o algoritmo do SMOTE adiciona uma “pessoa” intermediária medindo 1,81 m e pesando 78,3 kg. Vale notar que o SMOTE não seleciona necessariamente a média entre ambas as informações preexistentes. Segue aqui um exemplo visual:

![alt text](https://raw.githubusercontent.com/rafjaa/machine_learning_fecib/master/src/static/img/smote.png)

Essa função `transform` faz várias coisas na nossa base de dados, a primeira coisa é a transformação das classes do nosso problema em valores numéricos. Os algoritmos de Machine Learning, em sua grande maioria, trabalham com contas matemáticas e utilizar variáveis como _race_ não vai nos ajudar. Portanto vamos transformar as variáveis categóricas em numéricas, e como faremos isso? Por exemplo, vimos que na coluna _race_ temos 5 tipos de valores: White, Black, Asian, Armer Indian e Other, podemos associar agora cada instância de um desses valores como um número: White como 0, Black como 1 e assim por diante. Assim faremos para todas as variáveis categóricas. Esse método é chamado de **Label Encoder**.

Logo após isso teremos a fase onde aplicaremos o algoritmo SMOTE nos nossos dados. Depois normalizaremos toda a nossa base de dados com o desvio padrão. Ou seja, para cada coluna calcularemos o desvio padrão e a média e em cada instância iremos diminuir da sua média e dividir o resultado pelo desvio padrão. Isso irá evitar que o algoritmo dê pesos diferentes para uma classe de idade e a de saldo. Por fim separamos aqui os dados em treino e em teste para avaliar os nossos modelos

## Regressão Logística Multiclasse

![alt text](https://shapeofdata.files.wordpress.com/2013/06/multiava.png)
<br>
Regressão logística é um algoritmo que tenta prever o caso de uma classe binária, na qual tenta fazer com que oos dados se comportem como uma função 1/(1+e^-x). No nosso caso isso não seria suficiente e por isso usaremos a versão de Multiclasse. Faremos a classficicação utilizando da mesma estratégia da Regressão Logística no caso binário de pertencer ou não a uma determinada classe e faremos isso para todas as classes e todos os dados. A partir disso como a função 1/(1+e^-x) está entre 0 e 1 podemos interpretar como uma probabilidade, a probabilidade de um dado pertencer a uma classe. Com isso aplicaremos a função softmax e calcularemos qual a classe tem maior probabilidade e classificaremos cada dado assim.

```python
parametros = {
            'penalty': ['l2'],
            'C': [(i/10) for i in range(10,61,10)],
            'solver' : ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga'],
            'max_iter' : [10000]
        }
prev_LR,acc_LR = bestClassfier(X,Y,x,y,parametros,LogisticRegression,'Logistic',label,'f1',True)
```

## Random Forest
![alt text](https://miro.medium.com/max/1400/1*hmtbIgxoflflJqMJ_UHwXw.jpeg)
<br>
Random Forest é uma variação de outro algoritmo de Machine Learning, o Decision Tree, porém optei por colocar esse por ter uma performasse melhor. Ele irá fazer um cálculo para saber qual é a feature determinante para a escolha de classficação. Escolhido o feature, então a rede cria ramos para cada possibilidade que essa feature possibilita, por exemplo se essa feature for "Cor" poderia se separar entre "Vermelho","Verde","Azul", e a partir dessa separação irá se calcular qual a segunda principal feature para cada ramo que classifica os dados e assim criando outros ramos e se repetindo para todas as features selecionadas. A diferença do Random Forest para o Decision Tree é que neste vão todas as linhas e colunas do seu dataset de treino enquanto aquele é separado vários datasets menores (um parâmetro que pode ser alterdo no código) na qual tem linhas e colunas escolhidas aleatoriamente.

## KNN
![alt text](https://res.cloudinary.com/dyd911kmh/image/upload/f_auto,q_auto:best/v1531424125/KNN_final1_ibdm8a.png)
<br>
O KNN ou K-Nearst Neightbors (K-viznhos mais próximos) é um algoritmo na qual é utilizado a distância entre os dados para saber a sua classificação. O número K no nome indica quantos vizinhos você quer ser considerado, então no caso de K = 5 serão os 5 vizinhos mais próximos a serem utilizados para classificar o dado. Se os 5 forem de uma classe A, o dado será classificado como da Classe A, caso tenha 3 da Classe B 2 da A e 1 da C, será escolhido o que tiver maior número de classes no caso será classificado como a classe B.

```python
parametros = {
            'weights': ['uniform', 'distance'],
            'n_neighbors': [(int(i)) for i in range(3,11,2)],
            'algorithm' : ['auto', 'ball_tree', 'kd_tree', 'brute'],
            'p' : [1,2]
}
prev_KNN, acc_KNN = bestClassfier(X,Y,x,y,parametros,KNeighborsClassifier,'neighbors',label,'f1',True)
```

## Redes Neurais
![alt text](https://www.romsoc.eu/wp-content/uploads/image.png)
<br>
As Redes Neurais tentão simular como seria o aprendizado do nosso cérebro. Nessa estrutura, os padrões são introduzidos na rede neural pela camada de entrada e é comunicada a uma ou mais camadas ocultas presente na rede. As camadas ocultas recebem este nome somente por não constituírem a camada de entrada ou saída, são como camadas intermediárias. São nestas camadas que todo o processamento acontece por meio de um sistema de conexões dos chamados pesos e vieses. A entrada é recebida, o neurônio calcula uma soma ponderada adicionando também o viés e de acordo com o resultado e uma função de ativação predefinida, ele decide se deve ser 'disparado' ou ativado. Posteriormente, o neurônio transmite a informação para outros neurônios conectados em um processo chamado “forward pass”. Ao final desse processo, a última camada oculta é vinculada à camada de saída que possui um neurônio para cada saída possível desejada, tornando, assim, a previsão possível.

```python
parametros = {#
    #
    'activation': ['tanh',"identity",'logistic', 'relu'],
    'solver': ['adam'],
    'learning_rate': ['constant', 'invscaling', 'adaptive'],
    'tol' : [1e-4],
    'max_iter' : [1000],
    'hidden_layer_sizes' : [(8,4)]
}
prev_Neural, acc_Neural = bestClassfier(X,Y,x,y,parametros,MLPClassifier,'neural',label,'accuracy',True)
```

## Qual a medida foi utilizada?
Comparamos aqui vários resultados e discutimos qual foi o melhor modelo. Mas como sabemos se um resultado foi bom ou ruim? Na análise anterior utilizamos três parâmetros: resultados, tempo de execução e compreensão. Os dois últimos são bastante explicativos e não preciso dissertar tanto, entretanto o primeiro não. Se levarmos em conta somente a **acurácia** e jogarmos os dados isso não será suficiente, como as nossas classes são desbalanceadas conseguimos um resultado de 93% para uma, porém 40% para outra e isso é horrível. Por isso apelei para uma medida mais geral e visual: a **Matriz de confusão**. Ela dá origem a outros parâmetros como F1-score, revocação ou precisão.

Por fim, na hora de conseguir os melhores parâmetros, se for na documentação da função `bestClassfier` verá que eu passo um parâmetro chamado `score` para saber qual a melhor medida, foi utilizada os seguintes parâmetros: average_precision, accuracy,precision,f1,recall. Em geral eles não mudavam muito, a mudança era sentida na 3 ou 4 casa decimal depois da vírgula, mas no caso da Random Forest o melhor a ser utilizado foi a **Precisão**.
